### 4. Você deve ter percebido que as colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método pd.to_datetime(), converta ambas colunas para Timestamp / Datetime.

### 5. Crie uma nova coluna para representar o ano e o mês(aaaa-mm), utilizando a coluna DATA FINAL como referência.

### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
df2["MES"] = df2["DATA FINAL"].apply(lambda x: x.month) # Msm esquema das func anteriores.
df_rio = df2[df2["ESTADO"] == "RIO DE JANEIRO"] # Apos criar a coluna mes, filtrando apenas pelo estado do RJ
df_rio
```

```python

df_month_rio = df_rio.groupby("ANO-MES")[["PREÇO MÉDIO REVENDA", "MES"]].last() # Pega a ultima informaçao de cada mes
# df_month_rio # shift, ele faz um deslocamento do meu conjunto de dados, mantendo o meu indice fixo
# df_month_rio[df_month_rio["MES"] == 12].shift(1) #para entender melhor execute separado, e observe o resultado
# ele joga o valor do mes seguinte para 'baixo'
(df_month_rio[df_month_rio["MES"] == 12] / df_month_rio[df_month_rio["MES"] == 12].shift(1) - 1) * 100
# Resumo: o shift joga o valor uma linha abaixo, com isso conseguimos '/' dividir um ano pelo outro, obtendo a variação.
```

### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

##### Uma serie temporal, nada mais que é uma tabela que contem informaçoes de tempo, e mostrar os valores mais caros e mais baratos em um periodo X

```python
df2.groupby("ANO-MES").max()["PREÇO MÉDIO REVENDA"] # Isso ja é um serie temporal, que mostra os maiores valores de 'PREÇO MÉDIO REVENDA'
# No groupby, o 'ANO-MES' vira indice, no caso, as datas como '2004-05-01' 
```

```python
df_max = df2.groupby("ANO-MES").max()["PREÇO MÉDIO REVENDA"] # Os Maiores Valores 
df_min = df2.groupby("ANO-MES").min()["PREÇO MÉDIO REVENDA"] # Os Menores Valores

idx_max = df2.groupby('ANO-MES')["PREÇO MÉDIO REVENDA"].idxmax() # Retorna o inidice na qual o valor max foi detectado
idx_min = df2.groupby('ANO-MES')["PREÇO MÉDIO REVENDA"].idxmin() # Retorna o inidice na qual o valor min foi detectado
# Vou utilizar o indice p localizaçao no DF original, usar essa func para saber a posiçao dos maiores e menores valores

df_diff = pd.DataFrame() # DF vazio
```

```python
# idx_max
```

```python
df_max
```

```python
# Preenchendo o DF vazio 'df_diff' e Como os indices sao os msm, sao comum, ele os mantem, o ANO-MES
df_diff["abs_diff"] = df_max - df_min # Diferença absoluta
df_diff["perct_diff"] = (df_max - df_min) / df_min * 100 # Conta de percentual, p saber a diferença em % 
df_diff["max"] = df_max # Maior valor baseado no ano mes
df_diff["min"] = df_min # Menor valor baseado no ano mes
```

```python
# Agr a tabela mostra qual estado teve o maior e menor valor no mes.
df_diff["ESTADO_MIN"] = df2.loc[idx_min, :]["ESTADO"].values
df_diff
```
